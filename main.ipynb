```python
def make_dumb(df, column):
    df_copy = df.copy()
    classes = df_copy[column].unique()
    for c in classes:
        df_copy.loc[:, c] = df_copy[column] == c
    
    return df_copy
```

```python
df_copy = df.copy()

df_copy.loc[:, "Ram"] = [int(i[:-2]) for i in df_copy.loc[:, "Ram"]]
df_copy.loc[:, "Cpu Rate"] = [float(i[:-3]) for i in df_copy.loc[:, "Cpu Rate"]]
df_copy.loc[:, "Windows"] = [(i == "Windows 10") for i in df_copy.loc[:, "OpSys"]]
df_copy
```
